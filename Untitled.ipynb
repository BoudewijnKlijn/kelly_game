```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_capital(
    n_games = 1_000,
    n_turns_per_game = 1_000,
    p = 0.6, # 60% win chance
    bet_size = 0.2, # bet size as percentage of capital, 0.2 = 20%. 20% is optimum for this bet according to Kelly criterion
    start_capital = 100,
    win = 1, # +100% if you win
    lose = -1, # -100% if you lose
):
    assert 0 < bet_size <= 1, "Bet has to be: 0 < bet <= 1."
        
    outcomes = [win, lose]
    outcome_chances = [p, 1 - p]
    
    returns = 1 + bet_size * np.random.choice(
        outcomes, 
        size=(n_games, n_turns_per_game), 
        p=outcome_chances,
    )
    
    capital = start_capital * returns.cumprod(axis=1)
    
    return capital
```

```python
capital = simulate_capital()
```

```python
mean_capital_after_n_turns = capital.mean(axis=0)
median_capital_after_n_turns = np.median(capital, axis=0)
```

```python
plt.semilogy(mean_capital_after_n_turns, label='mean')
plt.plot(median_capital_after_n_turns, label='median')
plt.legend()
plt.show()
```

```python
log_capital = np.log(capital)
```

```python
mean_log_capital_after_n_turns = log_capital.mean(axis=0)
median_log_capital_after_n_turns = np.median(log_capital, axis=0)
```

```python
plt.plot(mean_log_capital_after_n_turns, label='mean')
plt.plot(median_log_capital_after_n_turns, label='median')
plt.legend()
plt.show()
```

```python
terminal_mean = list()
bet_sizes = np.linspace(0.01, 1, 20)

for bet_size in bet_sizes:
    capital = simulate_capital(bet_size=bet_size)
    log_capital = np.log(capital)
    mean_log_capital_after_n_turns = log_capital.mean(axis=0)
    terminal_mean.append(mean_log_capital_after_n_turns[-1])
    
plt.plot(bet_sizes, terminal_mean)
plt.show()
```
